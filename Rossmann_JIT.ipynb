```python
# Agregacja danych wg typu lokalizacji
grouped_loc = df.groupby("Typ lokalizacji").agg({
    "Średnia konwersja": "mean",
    "Wskaźnik satysfakcji klienta": "mean"
}).reset_index()
```

```python
# Wykres 1 – Średnia konwersja wg lokalizacji
plt.figure(figsize=(8,5))
plt.bar(grouped_loc["Typ lokalizacji"], grouped_loc["Średnia konwersja"], alpha=0.7)
plt.xticks(rotation=30, ha="right")
plt.ylabel("Średnia konwersja")
plt.title("Średnia konwersja wg typu lokalizacji")
plt.show()

```

```python
# Agregacja średniego koszyka wg typu lokalizacji
grouped_basket = df.groupby("Typ lokalizacji").agg({
    "Średni koszyk total [pln]": "mean"
}).reset_index()
```

```python
# Wykres – średni koszyk wg lokalizacji
plt.figure(figsize=(8,5))
plt.bar(grouped_basket["Typ lokalizacji"], grouped_basket["Średni koszyk total [pln]"], alpha=0.7, color="green")
plt.xticks(rotation=30, ha="right")
plt.ylabel("Średni koszyk [PLN]")
plt.title("Średni koszyk wg typu lokalizacji")
plt.show()

grouped_basket
```

```python
from pptx import Presentation
from pptx.util import Inches, Pt
import matplotlib.pyplot as plt

# Funkcja pomocnicza do tworzenia wykresów i zapisywania jako obraz
def save_plot_as_image(df, x_col, y_col, title, ylabel, filename, color="blue"):
    plt.figure(figsize=(7,4))
    plt.bar(df[x_col], df[y_col], color=color, alpha=0.7)
    plt.xticks(rotation=30, ha="right")
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(filename)
    plt.close()

# Zapisanie wykresów jako obrazki
save_plot_as_image(grouped_loc, "Typ lokalizacji", "Średnia konwersja",
                   "Średnia konwersja wg typu lokalizacji", "Średnia konwersja", "/mnt/data/konwersja.png")

save_plot_as_image(grouped_basket, "Typ lokalizacji", "Średni koszyk total [pln]",
                   "Średni koszyk wg typu lokalizacji", "Średni koszyk [PLN]", "/mnt/data/koszyk.png", color="green")

# Tworzenie prezentacji
prs = Presentation()

# Slajd tytułowy
slide_layout = prs.slide_layouts[0]
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[1]
title.text = "Analiza wskaźników klienckich drogerii"
subtitle.text = "Marzec 2024 – 300 lokalizacji"

# Slajd: Czynniki wpływające na konwersję
slide_layout = prs.slide_layouts[1]
slide = prs.slides.add_slide(slide_layout)
title, content = slide.shapes.title, slide.placeholders[1]
title.text = "Czynniki wpływające na konwersję"
content.text = (
    "- Negatywnie: wysoki traffic (więcej odwiedzin → niższa konwersja)\n"
    "- Pozytywnie: większa liczba sztuk na paragonie (r=0.48)\n"
    "- Pozytywnie: wyższy średni koszyk (r=0.22)\n"
    "➡ Konwersję wspiera bardziej zaangażowany klient z większymi zakupami"
)

# Slajd: Czynniki budujące satysfakcję klienta
slide = prs.slides.add_slide(slide_layout)
title, content = slide.shapes.title, slide.placeholders[1]
title.text = "Czynniki budujące satysfakcję klienta"
content.text = (
    "- Pozytywnie: skanowalność aplikacji (szczególnie samoobsługa, r≈0.24)\n"
    "- Neutralne: średni koszyk, liczba sztuk\n"
    "- Negatywne: wysoki traffic i liczba transakcji (lekkie obniżenie satysfakcji)\n"
    "➡ Aplikacja i samoobsługa budują pozytywne doświadczenie"
)

# Slajd: Trend 1 – Konwersja wg lokalizacji
slide_layout = prs.slide_layouts[5]
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
title.text = "Trend 1 – Średnia konwersja wg lokalizacji"
slide.shapes.add_picture("/mnt/data/konwersja.png", Inches(1), Inches(1.5), width=Inches(7))

# Slajd: Trend 2 – Średni koszyk wg lokalizacji
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
title.text = "Trend 2 – Średni koszyk wg lokalizacji"
slide.shapes.add_picture("/mnt/data/koszyk.png", Inches(1), Inches(1.5), width=Inches(7))

# Slajd końcowy – Podsumowanie
slide_layout = prs.slide_layouts[1]
slide = prs.slides.add_slide(slide_layout)
title, content = slide.shapes.title, slide.placeholders[1]
title.text = "Podsumowanie"
content.text = (
    "1. Konwersja rośnie wraz z większą liczbą produktów i wartością koszyka, spada przy dużym trafficu.\n"
    "2. Satysfakcję klienta zwiększa aplikacja i samoobsługa, a obniża tłok i pośpiech.\n"
    "3. Najlepsza konwersja: osiedla i ulice handlowe.\n"
    "4. Najwyższy koszyk: retail parki i małe centra handlowe; najniższy: dworce PKP."
)

# Zapis prezentacji
output_path = "/mnt/data/Analiza_wskaznikow_drogerii.pptx"
prs.save(output_path)

output_path
```
